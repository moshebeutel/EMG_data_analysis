# EMG Signal Features

There are 3 groups commonly used features:
1. Time Domain (TD) - The most commonly used. Contains:
    - Mean Absolute Value (MAV)
    - Zero Crossing (ZC)
    - Slope Sign Changes (SSC)
    - Waveform Length (WL)
2. Enhanced TD - Contains:
    - Skewness
    - Root Mean Square (RMS)
    - Integrated EMG (IEMG)
    - Autoregression Coefficients (AR)
    - Hjorth parameters
   In addition to TD features
3. Nina Pro Features - The to preform on the article presented the Nina Pro dataset. Contains:
    - RMS
    - Marginal Discrete Wavelet Transform (mDWT)
    - EMG Histogram
   In addition to TD features.
4. SampEn Pipeline - The most discriminate ones among a set fifty commonly used signal features.

In this notebook I will use the TD feature set with $window=256, step=128$
The target column **TRAG_GT_LAST** is the last value in window of  **TRAJ_GT**

```python
data = pd.read_hdf(f'{file_path}/{trial_name}.hdf5')
```

## MAV

```python
plt.figure(figsize=(25,10))
plt.plot(data.EMG_1)
wl_series = feature_wl(data.loc[:, 'EMG_1'], window=256, step=128)
plt.plot(wl_series)
```
